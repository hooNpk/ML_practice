```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
reg = LinearRegression()

Xsample=[[1],[2],[3],[4],[5]]
Ysample=[[3],[5],[7],[9],[11]]

plt.title('data points')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(Xsample,Ysample)
```

```python
Model=reg.fit(Xsample,Ysample)
print("coef")
print(Model.coef_)
print("intercept")
print(Model.intercept_)
```

```python
Model.predict([[15]])
```

```python
plt.title('Linear Regression sample')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(Xsample,Ysample)
plt.plot(Xsample,Model.coef_*Xsample + Model.intercept_)
```

```python
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()
X=california.data
DF=pd.DataFrame(X,columns=california.feature_names)
Y=california.target
print(DF)
```

```python
i=2
plt.title(california.feature_names[i]+' & ' + 'target')
plt.xlabel(california.feature_names[i])
plt.ylabel('target')
plt.scatter(DF[california.feature_names[i]],Y)
plt.legend()
plt.show()
```

```python
reg = LinearRegression()
Model=reg.fit(X,Y)
print("coef")
print(Model.coef_)
print("intercept")
print(Model.intercept_)
```

```python
i=0
plt.title(california.feature_names[i]+' & ' + 'target')
plt.xlabel(california.feature_names[i])
plt.ylabel('target')
plt.scatter(DF[california.feature_names[i]],Y)
plt.plot(DF[california.feature_names[i]],Model.coef_[i]*DF[california.feature_names[i]],'r-')
plt.legend()
plt.show()
```
